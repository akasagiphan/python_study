```python
# Thực hành vẽ mặt cười
import turtle

pen = turtle.Turtle()

pen.pensize(2)
pen.shape("square")
pen.penup()
pen.goto(0,-200)

facesize = 200
eyesize = 30
nosesize = 35

# Face
pen.pendown()
pen.fillcolor("yellow")
pen.begin_fill()
pen.circle(facesize)
pen.end_fill()

# Left eye
pen.penup()
pen.goto(-70, 70)

pen.pendown()
pen.fillcolor("black")
pen.begin_fill()
pen.circle(eyesize)
pen.end_fill()


# Right eye
pen.penup()
pen.goto(70, 70)

pen.pendown()
pen.fillcolor("black")
pen.begin_fill()
pen.circle(eyesize)
pen.end_fill()

# Nose
pen.penup()
pen.goto(0, 20)

pen.pendown()
pen.fillcolor("red")
pen.begin_fill()
pen.circle(-nosesize, steps = 3)
pen.end_fill()

# Mouth
pen.penup()
pen.goto(-100, -90)

pen.pendown()
pen.fillcolor("")
pen.right(60)
pen.begin_fill()
pen.circle(120, 120)
pen.end_fill()

turtle.done()
```

```python
import math

a = float(input("Nhập độ dài cạnh a: "))
b = float(input("Nhập độ dài cạnh b: "))
c = float(input("Nhập độ dài cạnh c: "))

s = (a+b+c)/2

area = math.sqrt(s*(s-a)*(s-b)*(s-c))

print("Diện tích hình tam giác là: ", area)
```
